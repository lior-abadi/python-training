```python
# Making a basic graph

# import Bokeh
from bokeh.plotting import figure 
from bokeh.io import output_file, show

# Prepare the data (same amount in x and y axis)
x = [1,2,3,4,5]
y = [6,7,8,9,10]

# Prepare the output file
output_file("Line.html")
# Create figure object
f = figure()

# Create line plot
f.line(x,y)

# Show object
show(f)
```

```python
# Making a basic graph

# import Bokeh and Pandas
from bokeh.plotting import figure 
from bokeh.io import output_file, show
import pandas

# Prepare the data (same amount in x and y axis)
df= pandas.read_csv("bachelors.csv")
x = df["Year"]
y = df["Engineering"]



# Create figure object
f = figure(plot_width=500,plot_height=400, tools='pan',logo=None)
f.title.text="Cool Data"
f.title.text_color="Gray"
f.title.text_font="times"
f.title.text_font_style="bold"
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Date"
f.yaxis.axis_label="Intensity"    

# Create line plot
f.line(x,y)

# Prepare the output file
output_file("BachelorsPlotEdited.html")

# Show object
show(f)
```
